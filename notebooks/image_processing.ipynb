```python
# Importing usefull libraries

import cv2 
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Downloading the image

!wget 'https://a.espncdn.com/photo/2023/0116/r1118398_1296x518_5-2.jpg'
```

```python
# Let us visulize the image

img = mpimg.imread('/content/arsenal.jpg')
print(img.shape)
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Let us reshape the image

img = Image.open('/content/arsenal.jpg')
resized_img = img.resize((600,600))
resized_img.save('/content/arsenal_resized.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Let us see the shape of our resized image

img = mpimg.imread('/content/arsenal_resized.jpg')
img.shape
```

```python
# Now let us convert this RGB image into gray scale

img = cv2.imread('/content/arsenal_resized.jpg')
gray_scale_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gray_scale_img.shape)
plt.imshow(gray_scale_img)
plt.axis('off')
plt.show()
```

```python
# Now let us save the gray scaled image

cv2.imwrite('arsenal_gray_scale.jpg',gray_scale_img)
```
